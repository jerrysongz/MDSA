# Timing Experiments

## Comparing Sorting Algorithms

### DATA 601

In this notebook, we'll compare two sorting algorithms:
_Selection sort_ (which is $O(n^2)$) and the built-in list [sort](https://docs.python.org/3.8/library/stdtypes.html#list.sort) available in Python (which is $O(n \lg n)$).

The notebook demonstrates how to use the `%timeit` line magic to time executions and collect timing data. Additionally, we'll also visualize the results for comparison. 

```python
# Collect all imports
import random
import math
import matplotlib.pyplot as plt
```

```python
# List sizes representing the input n
# increase MAX_POWER to control the size of the largest input
# to test against. Beware of memory limitations.
MAX_POWER = 10
N = [2**i for i in range(1,MAX_POWER)]
```

```python
# Built-in sort
# Time executions storing timing results
resultSort = []
for n in N:
    # Time execution by repeating a number of times using the timeit magic 
    # Store result so that we can use it later
    mylist = [random.random() for i in range(n)]
    res = %timeit -n 10 -o -q mylist.sort()
    resultSort.append(res)
```

```python
# Selection selection sort
def selectionSort( arr ):
    "Sort a list in-place using selection sort"
    for i in range(len(arr) - 1):
        least = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[least]:
                least = j
            
        #swap items at indices i and least
        tmp = arr[i]
        arr[i] = arr[least]
        arr[least] = tmp
    return arr
```

```python
# Time selection sort
resultSelSort = []
for n in N:
    # Time execution, only one execution, this is very slow! 
    # Store result so that we can use it later
    mylist = [random.random() for i in range(n)]
    res = %timeit -n 1 -o -q selectionSort( mylist )
    resultSelSort.append(res)
```

```python
# Plot average results

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.loglog( N, [res.average for res in resultSort] ,'k.-', label='Built-in Sort')
plt.loglog( N, [res.average for res in resultSelSort] ,'b.-', label='Selection Sort')
plt.legend();
fig.set_size_inches(8,6)
plt.show()
```

```python
# Determine slope for selection sort in log-log plot from the last
# two data points
( (math.log(resultSelSort[-1].average) - math.log(resultSelSort[-2].average) ) / 
 ( math.log(N[-1]) - math.log(N[-2]) ) )

```
