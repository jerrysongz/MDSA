As a percentage of all language speakers in each province or territory, how commonly is French spoken?

```python
create_statement1 = '''
CREATE TEMPORARY TABLE IF NOT EXISTS table1 AS (select GEO, `VALUE` from languages
WHERE `Knowledge of languages (324)` = 'Total - Knowledge of languages' 
AND `Single and multiple responses of knowledge of languages (3)` = 'Total - Single and multiple responses of knowledge of languages'
ORDER BY `VALUE` DESC);
'''

create_statement2 = '''
CREATE TEMPORARY TABLE IF NOT EXISTS table2 AS (select GEO, `VALUE` from languages
WHERE `Knowledge of languages (324)` = 'French' 
AND `Single and multiple responses of knowledge of languages (3)` ='Total - Single and multiple responses of knowledge of languages'
ORDER BY `VALUE` DESC);
'''

sql.execute(create_statement1,engine)
sql.execute(create_statement2,engine)
query1_table = pd.read_sql_query("SELECT * FROM table1", engine)
print(query1_table)
query2_table = pd.read_sql_query("SELECT * FROM table2", engine)
#print(query2_table)

create_statement3 = '''
select t1.GEO, (t2.`VALUE` / t1.`VALUE` * 100) as Percentagefrenchspeker
FROM table1 t1 INNER JOIN table2 t2
ON t1.GEO = t2.GEO 
ORDER BY Percentagefrenchspeker DESC;

'''

query3_table = pd.read_sql_query(create_statement3, engine)
print(query3_table)

#https://stackoverflow.com/questions/22082215/dividing-two-fields-from-two-different-tables
#https://stackoverflow.com/questions/30945513/how-to-divide-two-values-in-same-column
#help from leo hao su
```
