```r
bd <- function(k){
    p <- 1 -((choose(365,k) * factorial(k))/(365^{k}))
    print(p)
}

bd(10)
bd(5)
```

```r
kvalues = c(10,15,20,25,30,35,40,45,50,55)
yaxis <- bd(kvalues)
```

```r
plot(kvalues,yaxis,xlab="# of plp", main = "something", ylab = "prob 2 have same bd", type = "l", col = "red")
```
